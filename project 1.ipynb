17. Build a dataset with every possible (UserID, ProductID) pair (cross join)

```python
p_list = list(products["ProductID"])
```

```python
u_list = list(users["UserID"])
```

```python
p_list
```

```python
u_list
```

```python
import itertools
```

```python
combo = list(itertools.product(u_list, p_list))
```

```python
combo_df = pd.DataFrame(combo, columns = ["UserID", "ProductID"])
```

```python
combo_df.head()
```
